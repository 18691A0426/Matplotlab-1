# Matplotlib

# .pip3 install matplotlib

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

# Linear graph

```python
x=np.array([1,2,3])
y=np.array([5,7,4])
plt.plot(x,y,label="line")
plt.title("sample linear graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

# Histogram

. numerical data/quantitive

```python
x=[10,15,20,21,22,45,56,67,23,12,80]
y=[0,10,20,30,40,50,60,70,90,100]  #increasing order

plt.hist(x,y,histtype="bar",label="hist",rwidth=0.5)   #syntax
plt.xlabel("x hist")
plt.ylabel("y hist")

plt.title("histogram")
plt.legend()
plt.show()
```

# Pie chart

```python
a=np.array([10,3,4,12])
b=np.array(['green','blue','black','red'])
colors=['g','b','k','r']
plt.pie(a,labels=b,colors=colors,startangle=90)
plt.title("piechart")
plt.show()
```

# Scatter plot

```python
x1=np.array([5,8,10])
y1=np.array([12,6,6])
plt.scatter(x1,y1,label="scat",color='r')
plt.xlabel("x scat")
plt.ylabel("y scat")
plt.title("scatterplot")
plt.legend()
plt.show()
```

# Sample linear graph

```python
x1=np.array([1,2,3])
y1=np.array([5,7,4])
x2=np.array([6,9,10])
y2=np.array([4,3,8])
plt.plot(x1,y1,label='line one',color='b',linewidth=6)
plt.plot(x2,y2,label='line two',color='g',linewidth=6)
plt.title("sample linear graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

# Stack plot

```python
a=np.array([1,2,3,4,5])
b=np.array([2,3,4,3,2])
c=np.array([7,6,5,8,9])
d=np.array([8,7,5,4,10])
plt.plot(a,color='g',label="green",linewidth=5)
plt.plot(b,color='m',label='pink',linewidth=5)
plt.plot(c,color='b',label='blue',linewidth=5)
plt.plot(d,color='r',label='red',linewidth=5)
plt.stackplot(a,b,c,d,colors=['g','m','b','r'])       #syntax

plt.xlabel("x stack")
plt.ylabel("y stack")
plt.legend()
plt.show()
```
